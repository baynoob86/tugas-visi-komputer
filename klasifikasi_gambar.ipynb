```python
# ===== 1) Upload file foto tulisan angka =====
# Catatan: Baris ini hanya berfungsi di lingkungan Google Colab.
from google.colab import files
uploaded = files.upload()

# ===== 2) Utilitas Preprocess agar mirip MNIST (28x28, putih-di-atas-hitam) =====
import numpy as np
from PIL import Image, ImageOps

def preprocess_to_mnist_28x28(img_pil):
    """
    Fungsi ini melakukan pra-pemrosesan pada gambar PIL agar sesuai dengan format MNIST.

    Langkah-langkah:
    - Konversi ke grayscale
    - Auto-contrast
    - (Opsional) Invert warna jika latar belakang terang (agar digit jadi putih, latar jadi gelap seperti MNIST)
    - Crop ke bounding box digit
    - Resize dengan mempertahankan rasio ke (20x20), lalu pad ke (28x28)
    - Normalisasi ke rentang [0,1] dan tambah axis channel
    """
    # Grayscale + autocontrast
    img = img_pil.convert('L')
    img = ImageOps.autocontrast(img)

    arr = np.array(img).astype(np.uint8)
    # Jika rata-rata piksel terang (menandakan kertas putih), invert warnanya
    # supaya digit menjadi putih di atas latar gelap (gaya MNIST)
    if arr.mean() > 127:
        img = ImageOps.invert(img)
        arr = np.array(img)

    # Binarisasi ringan untuk mencari bounding box (kotak pembatas) digit
    thr = np.mean(arr) * 0.8  # Ambang batas adaptif sederhana
    mask = arr > thr

    # Hanya crop jika ada piksel yang terdeteksi di atas ambang batas
    if mask.any():
        ys, xs = np.where(mask)
        y0, y1 = ys.min(), ys.max()
        x0, x1 = xs.min(), xs.max()
        img = img.crop((x0, y0, x1 + 1, y1 + 1))

    # Resize ke 20x20 dengan mempertahankan aspect ratio
    img.thumbnail((20, 20), Image.Resampling.LANCZOS)
    w, h = img.size

    # Buat kanvas 28x28 hitam dan tempel gambar di tengah (padding)
    canvas = Image.new('L', (28, 28), color=0)
    canvas.paste(img, ((28 - w) // 2, (28 - h) // 2))

    # Normalisasi ke [0,1]
    arr = np.array(canvas).astype('float32') / 255.0
    # Tambah channel dimension agar menjadi (28, 28, 1)
    arr = arr[..., None]

    return canvas, arr
```

```python
from sklearn.metrics import accuracy_score

results = []
for fname in uploaded.keys():
    img_pil = Image.open(fname)
    _, x = preprocess_to_mnist_28x28(img_pil)   # x: (28,28,1) float [0,1]
    x_flat = x.reshape(1, -1)                   # (1,784)

    pred = int(clf.predict(x_flat)[0])

    # SVM default tidak punya proba kecuali SVC(probability=True). JikaAnda ingin probabilitas:
    # clf = svm.SVC(kernel='rbf', gamma='scale', probability=True) saatelatihan.
    conf = None
    try:
        if hasattr(clf, "predict_proba"):
            conf = float(np.max(clf.predict_proba(x_flat)))
    except Exception:
        pass

    results.append((fname, pred, conf))

    # Tampilkan hasil (gunakan gambar 28x28 yang sudah diproses di tahapCNN juga boleh)
    print(f"{fname} -> Prediksi SVM: {pred}" + (f" (p≈{conf:.2f})" if conf is not None else ""))
```
